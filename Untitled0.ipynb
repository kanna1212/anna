```
df = pd.read_csv(path + 'application_train.csv')
```

```
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 500)
df.isnull().sum()
```

```
X = df[['NAME_CONTRACT_TYPE','AMT_INCOME_TOTAL','NAME_FAMILY_STATUS','NAME_EDUCATION_TYPE']]
y = df[['TARGET']]



type_ohe1 = pd.get_dummies(X['NAME_FAMILY_STATUS'])
type_ohe2 = pd.get_dummies(X['NAME_EDUCATION_TYPE'])

X = pd.concat([X, type_ohe1], axis=1)
X.drop('NAME_FAMILY_STATUS', axis=1, inplace=True)

X = pd.concat([X, type_ohe2], axis=1)
X.drop('NAME_EDUCATION_TYPE', axis=1, inplace=True)

X.replace({'NAME_CONTRACT_TYPE': {'Cash loans': 0, 'Revolving loans': 1}}, inplace=True)



from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
X.head(5)
```

```
X.dtypes
```

```
lr = LogisticRegression(random_state=42)
lr.fit(X, y)

print('Logistic Regression \n')
print('Train Score: {}'.format(round(lr.score(X, y), 3)))
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=1)
clf = LogisticRegression()
clf.fit(X_train, y_train)
y_train_pred = clf.predict(X_train)
y_test_pred = clf.predict(X_test)

from sklearn.metrics import accuracy_score
train_score = accuracy_score(y_train, y_train_pred)
test_score = accuracy_score(y_test, y_test_pred)
```

```
train_score,test_score
```

```
# インポート
from sklearn.metrics import confusion_matrix

# テストデータを使って予測値を算出
y_pred = clf.predict(X_test)

m = confusion_matrix(y_test, y_pred)
print('Confution matrix:\n{}'.format(m))
```

```
accuracy = (m[0, 0] + m[1, 1]) / m.sum()
# 適合率の計算
precision = (m[1,1])/m[:, 1].sum()

# 再現率の計算
recall = (m[1,1])/m[1, :].sum()

# F1スコアの計算
f1 = 2 * (precision * recall)/(precision + recall)

print('正解率:{:.3f}'.format(accuracy))
print('適合率:{:.3f}'.format(precision))
print('再現率:{:.3f}'.format(recall))
print('F1値:{:.3f}'.format(f1))
```
